```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.datasets import *
from sklearn.linear_model import LogisticRegression 

data = load_breast_cancer()
```

```python
tuned_parameters = [{'C': [10**-4, 10**-2, 10**-0, 10**2, 10**4 ]}]
X_train, X_test, y_train, y_test = train_test_split(data.data, data.target, train_size = 0.9 )
```

```python
#using  GridSearchCV

model = GridSearchCV(LogisticRegression() , tuned_parameters , scoring = 'f1', cv =  5)    #estimator , param_grid, scoring, cv
model.fit(X_train , y_train)
  

    
print(model.best_estimator_)    #print LR = best lemda and remaining all parameters are default 

print(model.score(X_test, y_test))
```

```python
#using RandomizedSearchCV

from scipy.stats import uniform
c = uniform.rvs(loc = 0 , scale = 4)
print(c)
```

```python
from sklearn.model_selection import RandomizedSearchCV
X_train, X_test, y_train, y_test = train_test_split(data.data, data.target, train_size = 0.9 )

parameters = dict (C = uniform(loc =0 ,scale = 4), penalty =['l2'])

model = RandomizedSearchCV(LogisticRegression() , parameters , scoring = 'f1', cv =  5)    #estimator , param_grid, scoring, cv
model.fit(X_train , y_train)
  

    
print(model.best_estimator_)    #print LR = best lemda and remaining all parameters are default 

print(model.score(X_test, y_test))
```

```python
# More Sparsity (Fewer elements of W* being non-zero) by increasing Lambda (decreasing C) 

import numpy as np
                       # Lambda = 10
    
clf = LogisticRegression( C = 0.1 , penalty = 'l1',  solver='liblinear')
clf.fit(X_train, y_train)
w = clf.coef_               # w = weight vector classifier.cofficient_
print(np.count_nonzero(w))   # printing number of non zero
```
